```python
# Problem Statement:
# A social media platform wants to automatically detect abusive or toxic comments.
# Build a text classification model using embeddings and LSTM to classify text.

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# -------------------------------
# Step 1: Sample Data
# -------------------------------
texts = ["I love this", "This is bad", "Very nice", "Hate this"]
labels = [0, 1, 0, 1]   # 0 = normal, 1 = toxic

# -------------------------------
# Step 2: Tokenization
# -------------------------------
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

seq = tokenizer.texts_to_sequences(texts)
padded = pad_sequences(seq)

# -------------------------------
# Step 3: Build LSTM Model
# -------------------------------
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(100, 16),
    tf.keras.layers.LSTM(16),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# -------------------------------
# Step 4: Train
# -------------------------------
# Convert data to TensorFlow tensors
padded_tensor = tf.constant(padded, dtype=tf.int32)
labels_tensor = tf.constant(labels, dtype=tf.int32)

model.fit(padded_tensor, labels_tensor, epochs=10, verbose=0)

# -------------------------------
# Step 5: Prediction
# -------------------------------
test = ["This is horrible"]
test_seq = tokenizer.texts_to_sequences(test)
test_pad = pad_sequences(test_seq, maxlen=padded.shape[1])

pred = model.predict(test_pad)
print("Prediction:", pred)
```
