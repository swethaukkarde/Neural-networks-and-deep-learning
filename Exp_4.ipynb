```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import numpy as np
from PIL import Image

# Problem Statement:
# A wildlife rescue center needs a model to differentiate between images
# of cats, dogs, and birds. Build a CNN-based image classifier and
# explain how convolution layers help detect features like edges and shapes.

# -------------------------------
# Step 0: Ensure Data Directory Exists and Contains Dummy Images
# -------------------------------
data_dir = "data/"
classes = ["cat", "dog", "bird"]
num_dummy_images_per_class = 5

# Check if data directory already has content, if so, skip dummy creation
# This simple check avoids overwriting or recreating if user has already added data
has_content = False
if os.path.exists(data_dir):
    for class_name in classes:
        class_path = os.path.join(data_dir, class_name)
        if os.path.exists(class_path) and len(os.listdir(class_path)) > 0:
            has_content = True
            break

if not has_content:
    print(f"Creating dummy data directories and images in '{data_dir}'...")
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)

    for class_name in classes:
        class_path = os.path.join(data_dir, class_name)
        if not os.path.exists(class_path):
            os.makedirs(class_path)

        # Create dummy images (64x64 black RGB images)
        for i in range(num_dummy_images_per_class):
            dummy_image_path = os.path.join(class_path, f"{class_name}_{i:02d}.jpg")
            img_array = np.zeros((64, 64, 3), dtype=np.uint8)
            img = Image.fromarray(img_array)
            img.save(dummy_image_path)
    print(f"Created {num_dummy_images_per_class} dummy images for each class.")
else:
    print(f"'{data_dir}' already contains data. Skipping dummy data creation.")


# -------------------------------
# Step 1: Data Preparation
# -------------------------------
datagen = ImageDataGenerator(rescale=1./255)

train = datagen.flow_from_directory(
    "data/",              # folder with subfolders: cat, dog, bird
    target_size=(64,64),
    class_mode='categorical'
)

# -------------------------------
# Step 2: Build CNN Model
# -------------------------------
model = tf.keras.Sequential([
    tf.keras.Input(shape=(64,64,3)), # Recommended way to specify input shape for Sequential models
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')  # 3 classes
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# -------------------------------
# Step 3: Train Model
# -------------------------------
model.fit(train, epochs=5)

print("Model trained successfully")
```
