```
import numpy as np

vec = np.array([1, 2, 3, 4])
c = 2

print("Vector:", vec)
print("Scalar:", c)

print("Addition:", vec + c)
print("Subtraction:", vec - c)
print("Multiplication:", vec * c)
print("Division:", vec / c)
```
