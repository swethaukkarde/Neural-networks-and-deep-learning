```python
import numpy as np

# Create a vector
vec = np.array([1, 2, 3, 4])

# Define scalar value
c = 2

# Addition
add_op = vec + c
add_func = np.add(vec, c)

# Subtraction
sub_op = vec - c
sub_func = np.subtract(vec, c)

# Multiplication
mul_op = vec * c
mul_func = np.multiply(vec, c)

# Division
div_op = vec / c
div_func = np.divide(vec, c)

# Display results
print("Original Vector:", vec)
print("Scalar Value:", c)

print("\n--- Addition Results ---")
print("Addition (Operator):", add_op)
print("Addition (Function):", add_func)

print("\n--- Subtraction Results ---")
print("Subtraction (Operator):", sub_op)
print("Subtraction (Function):", sub_func)

print("\n--- Multiplication Results ---")
print("Multiplication (Operator):", mul_op)
print("Multiplication (Function):", mul_func)

print("\n--- Division Results ---")
print("Division (Operator):", div_op)
print("Division (Function):", div_func)
```
