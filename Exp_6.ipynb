```python
# Problem Statement:
# A network security team wants to detect unusual system activity.
# Build an LSTM Autoencoder to reconstruct normal patterns and
# explain how reconstruction error helps identify anomalies.

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, RepeatVector, TimeDistributed, Dense

# -------------------------------
# Step 1: Generate Sample Data
# -------------------------------
# Normal data (pattern)
data = np.array([[i, i+1, i+2] for i in range(50)], dtype=float)

# Normalize
data = data / np.max(data)

# Reshape for LSTM [samples, time steps, features]
data = data.reshape((data.shape[0], data.shape[1], 1))

# -------------------------------
# Step 2: Build LSTM Autoencoder
# -------------------------------
model = Sequential([
    LSTM(16, activation='relu', input_shape=(3,1), return_sequences=False),
    RepeatVector(3),
    LSTM(16, activation='relu', return_sequences=True),
    TimeDistributed(Dense(1))
])

model.compile(optimizer='adam', loss='mse')

# -------------------------------
# Step 3: Train Model
# -------------------------------
model.fit(data, data, epochs=50, verbose=0)

# -------------------------------
# Step 4: Reconstruction
# -------------------------------
reconstructed = model.predict(data)

# Calculate reconstruction error
error = np.mean(np.abs(data - reconstructed), axis=(1,2))

# -------------------------------
# Step 5: Detect Anomalies
# -------------------------------
threshold = 0.1
anomalies = error > threshold

print("Reconstruction Error:", error[:5])
print("Anomalies Detected:", anomalies[:5])
```
