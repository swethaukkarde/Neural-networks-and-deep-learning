```python
# Problem Statement:
# A customer service department wants to classify emails as complaint,
# request, or appreciation. Create a Transformer-based classifier and
# explain how self-attention identifies important words.

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np # Import numpy

# -------------------------------
# Step 1: Sample Data
# -------------------------------
texts = ["Need help", "Thank you", "Bad service"]
labels = [1, 2, 0]  # 0=complaint, 1=request, 2=appreciation

# -------------------------------
# Step 2: Tokenization
# -------------------------------
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

seq = tokenizer.texts_to_sequences(texts)
padded = pad_sequences(seq)

# -------------------------------
# Step 3: Transformer Model
# -------------------------------
inputs = tf.keras.Input(shape=(padded.shape[1],))

x = tf.keras.layers.Embedding(100, 16)(inputs)
x = tf.keras.layers.MultiHeadAttention(num_heads=2, key_dim=16)(x, x)
x = tf.keras.layers.GlobalAveragePooling1D()(x)

outputs = tf.keras.layers.Dense(3, activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# -------------------------------
# Step 4: Train
# -------------------------------
model.fit(padded, np.array(labels), epochs=10, verbose=0) # Convert labels to numpy array

# -------------------------------
# Step 5: Predict
# -------------------------------
test = ["Good service"]
test_seq = tokenizer.texts_to_sequences(test)
test_pad = pad_sequences(test_seq, maxlen=padded.shape[1])

pred = model.predict(test_pad)
print("Prediction:", pred)
```
